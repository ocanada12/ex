```python
# ### 여기서부터는 콘솔 입력창 ####
import pyautogui
from tkinter import *
import tkinter as tk
import numpy as np
import cv2

while True :
    try :
        list1 = []
        list2 = []

        window = Tk()

        window.title("자동 스크린샷 촬영")

        def pressed():
            label.configure(text="버튼을 누름")

        def shot1():
            in_text = "입력 내용 : " + input_text1.get()
            list1.append(input_text1.get())
            in_text = "입력 내용 : " + input_text2.get()
            list2.append(input_text2.get())
            pyautogui.screenshot(list1[0]+"_"+"P"+list2[0]+"_X10.png", region=(364, 145, 1500, 840))
            def bitOperation(hpos, vpos) :
                img1 = cv2.imread(list1[0]+"_"+"P"+list2[0]+"_X10.png")
                img2 = cv2.imread('C:/Users/Han Cheol Jin/Pictures/x10.png')    

                rows, cols, channels = img2.shape
                roi = img1[vpos:rows+vpos, hpos:cols+hpos]

                img2gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
                ret, mask = cv2.threshold(img2gray, 10, 255, cv2.THRESH_BINARY)
                mask_inv = cv2.bitwise_not(mask)

                img1_bg = cv2.bitwise_and(roi, roi, mask=mask_inv)

                img1[vpos:rows+vpos, hpos:cols+hpos] = img1_bg

                cv2.imwrite(list1[0]+"_"+"P"+list2[0]+"_X10.png",img1)

            bitOperation(1170,690)
            
        def shot2():
            in_text = "입력 내용 : " + input_text1.get()
            list1.append(input_text1.get())
            in_text = "입력 내용 : " + input_text2.get()
            list2.append(input_text2.get())
            pyautogui.screenshot(list1[0]+"_"+"P"+list2[0]+"_X20.png", region=(364, 145, 1500, 840))            
            def bitOperation(hpos, vpos) :
                img1 = cv2.imread(list1[0]+"_"+"P"+list2[0]+"_X20.png")
                img2 = cv2.imread('C:/Users/Han Cheol Jin/Pictures/x20.png')    

                rows, cols, channels = img2.shape
                roi = img1[vpos:rows+vpos, hpos:cols+hpos]

                img2gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
                ret, mask = cv2.threshold(img2gray, 10, 255, cv2.THRESH_BINARY)
                mask_inv = cv2.bitwise_not(mask)

                img1_bg = cv2.bitwise_and(roi, roi, mask=mask_inv)

                img1[vpos:rows+vpos, hpos:cols+hpos] = img1_bg

                cv2.imwrite(list1[0]+"_"+"P"+list2[0]+"_X20.png",img1)

            bitOperation(1170,690)
   
        def closing(event):
            window.destroy()
            
########################################################        

        input_text1 = Entry(window, width=10)
        input_text1.grid(column=0, row=2)
        input_text1.place(x=70,y=10)

        lb_1 = Label(window, text="Cell line", bg='gray19', fg='gray60',width=7, height=1)
        lb_1.place(x=10,y=10)

########################################################        
        
        input_text2 = Entry(window, width=10)
        input_text2.grid(column=0, row=2)
        input_text2.place(x=70,y=50)

        lb_2 = Label(window, text="Passage", bg='gray19', fg='gray60',width=7, height=1)
        lb_2.place(x=10,y=50)
        
########################################################        

        button = Button(window, text="X10", command=shot1)
        button.grid(column=1, row=2)
        button.place(x=10,y=80)

        button = Button(window, text="X20", command=shot2)
        button.grid(column=1, row=2)
        button.place(x=45,y=80)
        
########################################################
        
        lb_hi = Label(window, text="종료", bg='gray19', fg='gray60',width=5, height=1)
        lb_hi.bind('<Button-1>',closing)
        lb_hi.place(x=10,y=120)
        
########################################################        

        window.geometry('200x150-500+500')
        window.mainloop()

########################################################        
        
        ### 콘솔입력창 ###
        print(list1[0])
        print(list2[0])
    except :
        print("촬영이 끝났습니다.")
        break
        
```
